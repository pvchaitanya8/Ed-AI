## Dynamic Files - Reset

### Practice // MCQ Page

```python
import json

json_file = r'../dynamic files/Main_pages/Practice_MCQ.json'
with open(json_file, 'r') as file:
    data = json.load(file)

for key in data:
    data[key]["Completed"] = False

with open(json_file, 'w') as file:
    json.dump(data, file, indent=4)

print(json.dumps(data, indent=4))
```
